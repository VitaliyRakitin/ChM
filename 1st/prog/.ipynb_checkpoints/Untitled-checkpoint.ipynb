```python
import pandas as pd
import numpy as np
import pylab as pl
%pylab inline
read_df = pd.read_csv("data_-7.csv",sep = ",")

t = read_df[[0]].values
x = read_df[[1]].values
y = read_df[[2]].values

pl.plot(t, x)
pl.axis([0, 30 , -9 , 9])
pl.xlabel("t")
pl.ylabel("x")
savefig('t-x.pdf')

pl.show()
```

```python
pl.plot(t,y)
pl.axis([0, 30, -9 , 9])
pl.xlabel("t")
pl.ylabel("y")
savefig('t-y.pdf')

pl.show()
```

```python
pl.plot(x,y)
pl.axis([-9, 9, -9 , 9])
pl.xlabel("x")
pl.ylabel("y")
savefig('x-y.pdf')

pl.show()
```
